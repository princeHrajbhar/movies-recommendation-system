```python
import pandas as pd

movies = pd.read_csv(
    r"C:\Users\Prince\Desktop\New folder (3)\PYTHON PROJECTS\machine_learning_project\tmdb_dataset\tmdb_5000_movies.csv"
)

credits = pd.read_csv(
    r"C:\Users\Prince\Desktop\New folder (3)\PYTHON PROJECTS\machine_learning_project\tmdb_dataset\tmdb_5000_credits.csv"
)
```

```python
movies=movies.merge(credits,on='title')
```

```python
# genres
# id
# keywords
# title
# overview
# populaity
# production company
# cost
# crew
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies = movies.copy()
movies.dropna(inplace=True)
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)

```

```python
movies.isnull().sum()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    try:
        return [i['name'] for i in ast.literal_eval(obj)[:3]]
    except:
        return []
```

```python
movies.iloc[0].crew
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.isnull().sum()
```

```python
def fetch_director(obj):
    try:
        for i in ast.literal_eval(obj):
            if i['job'] == 'Director':
                return [i['name']]
        return []
    except:
        return []
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.isnull().sum()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.isnull().sum()
```

```python
movies['genres']   = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast']     = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew']     = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.isnull().sum()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
  new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(
    list(enumerate(similarity[0])),
    reverse=True,
    key=lambda x: x[1]
)[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]

    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
